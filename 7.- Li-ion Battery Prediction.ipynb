# Load dataset, Feature Filtering

```python
# Import needed libraries for this work to be replicated: {sklearn: v1.1.1/ numpy: v1.20.1/ seaborn: v0.11.1/ pandas: 1.2.4}

import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

```

```python
df = pd.read_csv(r"") #read the csv file with the specific features that have been generated by Magpie {in our case the file name is "3.- Initial Dataset + features.csv"}

df_materials = df['material'] #save the materials names that have been used
df1 = df.drop(["material", "Cond", "cond(log)"], axis=1) #dropping columns that are not features

y = df["Cond"] #saving the ionic conductivity of the materials in our initial dataset

#Variance threshold to get rid of constant and quasi-constant features with a threshold of 0.01
vt = VarianceThreshold(threshold=0.01)
vt.fit(df1)
vt.get_support()
concol = [column for column in df1.columns
           if column not in df1.columns[vt.get_support()]]
df2 = df1.drop(concol, axis=1)
df2= df2.dropna()
df2_columns = df2.columns #"df2" in this case is the dataframe containing the filtered features (109 features)
```

# Train Test Split

```python
#designating X and y as variables for the features and ionic condctivity respectively
X = df2
y = df["Cond"]
X_col = df2.iloc[:,:108]
y_col = df2.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1) #Split made for 70% training and 30% testing

#initiating and fitting the model with a random_state of 42
model = RandomForestRegressor(random_state=42,n_estimators=100, max_depth = None, min_samples_leaf= 1) # the hyperparameters have been optimized according to a Grid Search technique described in the supplementary materials
model.fit(X_train, y_train)
yhat = model.predict(X_test)
mae = mean_absolute_error(y_test, yhat)
mse = mean_squared_error(y_test, yhat)
ypred = model.predict(X_test)

#visualizing the results
sns.scatterplot(x=y_test,y= ypred) #Scatter plot of all predictions
```

```python
full_model = model.fit(X,y) #fit the model on all of the dataset (training and testing)
```

# Prediction

```python
df_validation = pd.read_csv(r"") #Load file with unseen/new compounds to be predicted, must have generated features,in our work the filename is "6.- Compound validation.csv"
df_valid = df_validation[X.columns] #Prediction of new unseen/untested compounds
y_validation = df_validation['Cond']
```

```python
y_pred_valid = full_model.predict(df_valid) #predict the validation set IC
```
